```python
#######################
#### Configuration ####
#######################

## Region to analyze ##
region = 12 # <<<----- Use number (11 to 24) based on table above

## Variable ##
## Select the variable to analyze from the list above
var = 'SST' # <<<----- Use short name given above. upper or lower case accepted.

## Period to analize ##
## Specify the period using the format: YYYY-MM-DD
## Data available specified above
## All data in monthly resolution
initial_date = '1998-01-01'
final_date   = '2017-08-31'

##############################
#### End of configuration ####
##############################
```

```python
def analyze_PICES_LME(lme,var,initial_date,final_date,timeres):
    
    %matplotlib inline  
    import sys
    sys.path.append('./subroutines/')
    from pices import get_pices_data
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import warnings
    warnings.simplefilter('ignore') # filter some warning messages
    from statsmodels.formula.api import ols
    lmenames = ['California Current','Gulf of Alaska','East Bering Sea','North Bering Sea','Aleutian Islands','West Bering Sea',
            'Sea of Okhotsk','Oyashio Current','Sea of Japan','Yellow Sea','East China Sea','Kuroshio Current',
            'West North Pacific','East North Pacific']

    # check values


    # assign variables
    lmei = lme
    lmename = lmenames[lmei-11]

    # data aquisition
    dtmean, dtclim, dtanom = get_pices_data(var, lmei, initial_date,final_date)
    nvar = 'sst'
    units = dtmean[nvar].attrs['units']
    #print(dtmean.sst.attrs)
    allvars = dtmean.data_vars
    
    varnamelong = dtmean[nvar].attrs['long_name']
    datasetname = dtmean[nvar].attrs['dataset']
    varname = nvar.upper()
    
    ## Data information
    print('Region = '+str(lmei)+' - '+lmename)
    print('Data = '+varnamelong)
    print('Units = '+units)
    print('Dataset = '+datasetname)
    print('Period = '+initial_date+' : '+final_date)
    
    # displaying time series data
    plt.figure(figsize=(10,4))
    plt.plot(dtmean.time,dtmean.sst)
    plt.grid(True)
    plt.ylabel(varname+' ('+units+')')
    plt.title(lmename+' '+varname+' values')
    plt.show()

    # display climatology
    plt.figure(figsize=(5,4))
    plt.plot(dtclim.month, dtclim.sst,'+-',color='k')
    plt.grid(True)
    plt.ylabel(varname+' ('+units+')')
    plt.xticks(range(1,13),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],rotation=45)
    plt.title(lmename+' '+varname+' climatology')
    plt.show()

    # display density plots
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.distplot(dtmean.sst, hist=True, kde=True, bins=30,
                 kde_kws={'linewidth': 2})
    plt.title(varname+' values density plot')
    plt.grid(True)
    plt.xlabel(varname+' ('+units+')')
    plt.subplot(1,2,2)
    sns.distplot(dtanom.sst, hist=True, kde=True, bins=30,
                 kde_kws={'linewidth': 2})
    plt.title(varname+' anomalies density plot')
    plt.grid(True)
    plt.xlabel(varname+' ('+units+')')
    plt.show()
    
    ## display statistics
    print('Mean value = ', round(dtmean.sst.values.mean(),2),units)
    print('Median value = ', round(np.median(dtmean.sst.values),2),units)
    print('Standard deviation = ', round(dtmean.sst.values.std(),2),units)
    print('\n')
    print('Maximum value = ', round(dtmean.sst.values.max(),2),units)
    print('Minimum value = ', round(dtmean.sst.values.min(),2),units)
    print('\n')
    print('Maximum anomalies value = ', round(dtanom.sst.values.max(),2),units)
    print('Minimum anomalies value = ', round(dtanom.sst.values.min(),2),units)
    
    # display anomalies
    plt.figure(figsize=(12,4),dpi=200)
    p=dtanom.where(dtanom>=0)
    n=dtanom.where(dtanom<0)
    plt.bar(p.time.values,p.sst, width=30, color='darkred',alpha=0.8, edgecolor=None,zorder=2)
    plt.bar(n.time.values,n.sst, width=30, color='darkblue',alpha=0.8, edgecolor=None,zorder=3)
    plt.grid(True,zorder=1)
    plt.ylabel(varname+' ('+units+')')
    plt.title(lmename+' '+varname+' anomalies')
    plt.autoscale(enable=True, axis='x', tight=True)
    #plt.savefig('GoA_SST.png')
    plt.show()
```

```python
#### do not modify ####

analyze_PICES_LME(lme, var, initial_date, final_date, timeres)
```
